## Iris Dataset

dataset source : https://www.kaggle.com/datasets/uciml/iris

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import random

# import python files
from hs_stats import z_score
```

```python
data = pd.read_csv('datasets/iris.csv')
display(data)
```

```python
# lists
SL = data["SepalLengthCm"]
SW = data["SepalWidthCm"]
PL = data["PetalLengthCm"]
PW = data["PetalWidthCm"]
Spec = data["Species"]
```

***
## Demonstration of Z-Score and Standard Deviation relationship using Sepal Lengths data

$datapoint = \bar{x} + (z \times \sigma)$

```python
print(f'min : {min(SL)}  max : {max(SL)}\n')

SLmean = round(float(SL.mean()),3)
SLvar = round(float(SL.var(0)),3)
SLstd = round(float(SL.std(0)),3)

print(f'mean : {SLmean}')
print(f'variance : {SLvar}')
print(f'standard deviation : {SLstd}')
```

```python
# random sepal lengths
rand_SL = [4.341, 6.35, 7.067]
```

```python
for x in rand_SL :
    print(f'SEPAL LENGTH {x}')
    z = round(z_score(SL,x),3)
    print(f' - z-score : {z}')
    print(f' - {SLmean} + ({z} * {SLstd}) = {round(SLmean + (z * SLstd),3)}\n')
```
